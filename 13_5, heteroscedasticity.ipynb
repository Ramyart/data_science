```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('13_5, heteroscedasticity.csv')
```

```python
data['log_B'] = np.log(data['Budget'])
data['log_V'] = np.log(data['Visits'])
```

```python
plt.scatter(data['Budget'] , data['Visits'])
plt.xlabel('Budget of Company')
plt.ylabel('Visits of Company')
```

```python
plt.scatter(data['log_B'] , data['Visits'])
plt.xlabel('Log(Budget of Company)')
plt.ylabel('Visits of Company')
```

```python
plt.scatter(data['Budget'] , data['log_V'])
plt.xlabel('Budget of Company')
plt.ylabel('Log(Visits of Company)')
```

```python
plt.scatter(data['log_B'] , data['log_V'])
plt.xlabel('Log(Budget of Company)')
plt.ylabel('Log(Visits of Company)')
```
